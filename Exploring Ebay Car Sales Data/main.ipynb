#### Converting the column names from camelcase to snakecase

```python
autos.columns
```

```python
# gonna identify letters wich are in upper case as division of words using ord() function
old_columns = autos.columns
new_columns = []
for col in old_columns:
    indexes_to_split = []
    words = []
    col = col.replace("Of", '')
    for i in range(len(col)):
        char = col[i]
        if ord(char) >= ord('A') and ord(char) <= ord('Z'):
            indexes_to_split.append(i)
    if col == 'powerPS':
        new_columns.append('power_ps')
    elif len(indexes_to_split) == 0:
        new_columns.append(col)
    else:
        for i in range(len(indexes_to_split)):
            if i == 0:
                words.append(col[:indexes_to_split[i]])
                if len(indexes_to_split) == 1:
                    words.append(col[indexes_to_split[i]:].lower())
                else:
                    words.append(col[indexes_to_split[i]:indexes_to_split[i + 1]].lower())
            elif i == (len(indexes_to_split) - 1):
                words.append(col[indexes_to_split[i]:].lower())
            else:
                words.append(col[indexes_to_split[i]:indexes_to_split[i + 1]].lower())
        new_columns.append('_'.join(words))

autos.columns = new_columns
for i in range(len(old_columns)):
    print(old_columns[i], new_columns[i])

# better suited columns
autos = autos.rename(columns={'year_registration' : 'registration_year', 'month_registration' : 'registration_month', 
'not_repaired_damage' : 'unrepaired_damage', 'date_created' : 'ad_created', 'kilometer' : 'odometer_km'})

print('\n', autos.columns)
```

Now, I'll calculate mean price for the 20 most common brands on the site

```python
brand_mean_prices = {}
brand_mean_mileages = {}
brand_mean_power = {}

top20_brands = autos['brand'].value_counts(normalize=True, sort=True, ascending=False).head(20).index
for brand in top20_brands:
    brand_bool = autos['brand'] == brand
    brand_prices = autos.loc[brand_bool, 'price_eur']
    brand_mileages = autos.loc[brand_bool, 'odometer_km']
    brand_power = autos.loc[brand_bool, 'power_ps']
    brand_mean_prices[brand] = brand_prices.mean(axis=0)
    brand_mean_mileages[brand] = brand_mileages.mean(axis=0)
    brand_mean_power[brand] = brand_power.mean(axis=0)

brand_mean_prices
```

```python
# bpm means brand price mean
# bpsm means brand ps mean
# bmm means brand mileage mean 
bpm_series = pd.Series(brand_mean_prices)
bpsm_series = pd.Series(brand_mean_power)
bmm_series = pd.Series(brand_mean_mileages)
brands_infos_df = pd.DataFrame(data = {'mean_price_eur' : bpm_series, 'mean_power_ps' : bpsm_series, 'mean_mileage_km' : bmm_series}).sort_values(by='mean_price_eur', ascending=False)
brands_infos_df.index.names=['brand']
brands_infos_df
```
